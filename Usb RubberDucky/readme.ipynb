# Usb RubberDucky

Penjelasan ini membantu memahami setiap langkah yang diambil oleh skrip, termasuk pengumpulan informasi sistem, pembentukan permintaan ke API GPT, dan pembuatan laporan dalam format HTML.

```python
REM Pembuat:YogaIT " 
REM Open PowerShell
DELAY 500                    ; Tunggu 500 milidetik
GUI x                        ; Tekan tombol Windows + X untuk membuka menu kontekstual
DELAY 2000                   ; Tunggu 2000 milidetik
STRING a                     ; Ketik "a" untuk memilih opsi PowerShell (Admin)
DELAY 2000                   ; Tunggu 2000 milidetik
LEFTARROW                    ; Tekan tombol panah kiri (untuk memastikan opsi dipilih)
DELAY 2000                   ; Tunggu 2000 milidetik
ENTER                        ; Tekan Enter untuk membuka PowerShell sebagai administrator

REM Membuat skrip PowerShell untuk mengumpulkan informasi sistem
DELAY 500                    ; Tunggu 500 milidetik
STRING $system_info = @{     ; Inisiasi hashtable untuk menyimpan informasi sistem
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'OS' = $(Get-CimInstance Win32_OperatingSystem).Caption;          ; Ambil nama OS
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'Version' = $(Get-CimInstance Win32_OperatingSystem).Version;      ; Ambil versi OS
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'Architecture' = $(Get-CimInstance Win32_OperatingSystem).OSArchitecture; ; Ambil arsitektur OS
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'ComputerName' = $(Get-CimInstance Win32_OperatingSystem).CSName;   ; Ambil nama komputer
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'LastBootTime' = $(Get-CimInstance Win32_OperatingSystem).LastBootUpTime; ; Ambil waktu boot terakhir
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'InstalledUpdates' = $(Get-HotFix | Sort-Object -Property InstalledOn -Descending | Select-Object -First 5).Description; ; Ambil deskripsi update yang terinstal
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'NetworkInfo' = $(Get-CimInstance Win32_NetworkAdapterConfiguration | Where-Object {$_.IPEnabled -eq $true}).IPAddress; ; Ambil informasi jaringan (IP address)
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'FirewallStatus' = $(Get-NetFirewallProfile | Where-Object { $_.Enabled -eq $true }).Name; ; Ambil status firewall yang aktif
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'UserAccounts' = $(Get-LocalUser | Where-Object { $_.Enabled -eq $true }).Name; ; Ambil nama akun pengguna yang aktif
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING 'RunningProcesses' = $(Get-Process | Sort-Object -Property CPU -Descending | Select-Object -First 5).Name; ; Ambil nama proses yang sedang berjalan dengan penggunaan CPU tertinggi
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING }                     ; Akhiri hashtable
ENTER

REM Meminta GPT untuk memformat respons dalam HTML
DELAY 500                    ; Tunggu 500 milidetik
STRING $prompt_text = "Given the detailed system information: OS: $($system_info.OS), Version: $($system_info.Version), Architecture: $($system_info.Architecture), Computer Name: COMPUTER_NAME_PLACEHOLDER, Last Boot Time: $($system_info.LastBootTime), Installed Updates: $($system_info.InstalledUpdates), Network Info: NETWORK_INFO_PLACEHOLDER, Firewall Status: $($system_info.FirewallStatus), User Accounts: USER_ACCOUNTS_PLACEHOLDER, Running Processes: $($system_info.RunningProcesses), provide a pentesting report identifying potential vulnerabilities in English, formatted in HTML with headers and bullet points for recommendations." ; Siapkan teks permintaan untuk API GPT
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING $messages = @(        ; Inisiasi array pesan untuk API GPT
ENTER
DELAY 1000                   ; Tunggu 1000 milidetik
STRING @{ 'role' = 'system'; 'content' = 'You are analyzing detailed system information for potential vulnerabilities.' }, ; Tambah pesan sistem untuk API GPT
ENTER
DELAY 1000                   ; Tunggu 1000 milidetik
STRING @{ 'role' = 'user'; 'content' = $prompt_text } ; Tambah pesan pengguna dengan teks permintaan
ENTER
DELAY 1000                   ; Tunggu 1000 milidetik
STRING )                     ; Akhiri array pesan
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING $headers = @{ 'Authorization' = 'Bearer YOUR_OPENAI_API_KEY'; 'Content-Type' = 'application/json' } ; Siapkan header untuk permintaan API GPT
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING $response = Invoke-RestMethod -Uri 'https://api.openai.com/v1/chat/completions' -Method POST -Headers $headers -Body (@{ model = 'gpt-3.5-turbo'; messages = $messages } | ConvertTo-Json) ; Kirim permintaan ke API GPT dan simpan respons
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING $htmlContent = @"     ; Mulai definisi string HTML
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING <html>                ; Buka tag HTML
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING <head>                ; Buka tag head
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING <title>Laporan Perentasan USB Autorun-GPT</title> ; Tambahkan judul halaman
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING <style>               ; Buka tag style
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING body {font-family: Arial, sans-serif; margin: 40px;} ; Tambahkan gaya untuk body
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING h2 {color: #333; border-bottom: 2px solid #eee; padding-bottom: 10px;} ; Tambahkan gaya untuk h2
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING h3 {color: #555; margin-top: 20px;} ; Tambahkan gaya untuk h3
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING p, ul {margin-bottom: 20px;} ; Tambahkan gaya untuk paragraf dan list
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING </style>              ; Tutup tag style
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING </head>               ; Tutup tag head
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING <body>                ; Buka tag body
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING <h2>Pentesting Report</h2> ; Tambahkan header utama
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING $($response.choices[0].message.content) ; Masukkan konten respons dari API GPT
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING </body>               ; Tutup tag body
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING </html>               ; Tutup tag HTML
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING "@                    ; Akhiri definisi string HTML
ENTER
DELAY 500                    ; Tunggu 500 milidetik
STRING Set-Content -Path $env:USERPROFILE\Desktop\Pentesting_Report.html -Value $htmlContent ; Simpan konten HTML ke file di desktop
ENTER

REM Mengganti tempat penampung dengan nilai aktual dalam laporan lokal
DELAY 500                    ; Tunggu 500 milidetik
STRING (Get-Content $env:USERPROFILE\Desktop\Pentesting_Report.html).Replace('COMPUTER_NAME_PLACEHOLDER', $system_info.ComputerName).Replace('NETWORK_INFO_PLACEHOLDER', ($system_info.NetworkInfo -join ', ')).Replace('USER_ACCOUNTS_PLACEHOLDER', ($system_info.UserAccounts -join ', ')) | Set-Content $env:USERPROFILE\Desktop\Pentesting_Report.html ; Ganti placeholder dengan nilai aktual dan simpan kembali file HTML
ENTER
```
